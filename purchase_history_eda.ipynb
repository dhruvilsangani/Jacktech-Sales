```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_20_26 = pd.read_excel("Pur_His.xlsx", sheet_name="20-26")
df_14_20 = pd.read_excel("Pur_His.xlsx", sheet_name="14-20")
```

```python
len(df_20_26)
```

```python
len(df_14_20)
```

```python
INVOICE_DATE_COL = "Invoice Date"

df_purchase_history = pd.concat([df_20_26, df_14_20], ignore_index=True)
df_purchase_history = df_purchase_history.drop_duplicates()
df_purchase_history[INVOICE_DATE_COL] = pd.to_datetime(
    df_purchase_history[INVOICE_DATE_COL], errors="coerce"
)
df_purchase_history = (
    df_purchase_history.sort_values(INVOICE_DATE_COL, na_position="last")
    .reset_index(drop=True)
)
df_purchase_history
```

```python
df_purchase_history.to_excel("df-purchase.xlsx", index=False)
```
